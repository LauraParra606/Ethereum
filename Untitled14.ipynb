```python
# -*- coding: utf-8 -*-
"""Implementaci√≥n.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1At4hb87Xri3Qra15lYXf4KLvj6xYe43f
"""

!pip install schedule
from google.colab import drive
drive.mount('/content/drive')

import schedule
import time
import pandas as pd
import requests
from datetime import datetime, timedelta
from sklearn.linear_model import LinearRegression
import numpy as np
import os

def main():
    data_file = '/content/drive/MyDrive/ethereum_data.csv'
    pred_file = '/content/drive/MyDrive/ethereum_predictions.csv'

    def get_ethereum_price():
        url = 'https://api.coingecko.com/api/v3/simple/price?ids=ethereum&vs_currencies=usd'
        response = requests.get(url)
        data = response.json()
        price = data['ethereum']['usd']
        time_now = datetime.now()
        return {'timestamp': time_now, 'price': price}

    new_row = pd.DataFrame([get_ethereum_price()])

    if os.path.exists(data_file):
        new_row.to_csv(data_file, mode='a', header=False, index=False)
    else:
        new_row.to_csv(data_file, mode='w', header=True, index=False)

    df = pd.read_csv(data_file)
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['timestamp_ordinal'] = df['timestamp'].map(datetime.toordinal)

    X = df[['timestamp_ordinal']]
    y = df['price']
    model = LinearRegression()
    model.fit(X, y)

    next_time = df['timestamp'].max() + timedelta(hours=4)
    next_time_ord = next_time.toordinal()
    predicted_price = model.predict(pd.DataFrame({'timestamp_ordinal': [next_time_ord]}))

    prediction_df = pd.DataFrame([{
        'timestamp': next_time,
        'predicted_price': predicted_price[0]
    }])

    if os.path.exists(pred_file):
        prediction_df.to_csv(pred_file, mode='a', header=False, index=False)
    else:
        prediction_df.to_csv(pred_file, mode='w', header=True, index=False)

    print(f"Script ejecutado a las {datetime.now().strftime('%H:%M:%S')}")

# Programar y ejecutar el script cada minuto
schedule.every(1).minutes.do(main)

while True:
    schedule.run_pending()
    time.sleep(1)
```
